### SVD - NMF

A non-convex solution space is characterized by multiple local optima and a complex landscape, making it challenging to find the global optimum. This is the case with NMF, where the non-negativity constraints and the nature of matrix multiplication create a non-convex optimization problem. In contrast, SVD operates within a convex solution space, leading to a unique, well-defined solution.

A 2D objective function that illustrates a non-convex solution space:
$$ f(x, y) = (x^2 - 4)^2 + (y^2 - 1)^2 $$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the objective function
def objective_function(x, y):
    return (x**2 - 4)**2 + (y**2 - 1)**2

# Generate data points for the plots
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)
Z = objective_function(X, Y)

# Create the contour plot
plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, levels=50)
plt.title('Contour Plot of the Objective Function')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()

# Create the surface plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Surface Plot of the Objective Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```
